## 한국 은행 총재 기자 간담회 크롤링

```python
import requests
import datetime
import pandas as pd
from bs4 import BeautifulSoup

url= "https://www.bok.or.kr/portal/bbs/B0000169/list.do?menuNo=200059&pageIndex={}"
```

```python
href_list= []

for i in range(1, 20): #, 20
    링크 = BeautifulSoup(requests.get(url.format(i), 'html.parser').text,'lxml').select('#content > div.photoList.type1 > ul > li > a')
    
    for i in 링크:
        href = i.attrs['href']        
        href_list.append('https://www.bok.or.kr/'+href)
# href_lists
    
```

```python
fin_result = []
for i in href_list:
    get_conf = BeautifulSoup(requests.get(i, 'html.parser').text,'lxml').select('#board > div.col.l12.s12 > div')
    get_date = BeautifulSoup(requests.get(i, 'html.parser').text,'lxml').select_one('#board > div.viewLine.bLine > div > span.dataInfo > span.date').get_text()
    
    for conf_arrange in get_conf:
        fin_result.append({
            'DATE' : pd.to_datetime(datetime.datetime.strptime(get_date.strip('등록일'), '%Y.%m.%d')),
            'CONTENTS': ','.join(conf_arrange.text.strip().splitlines())
            })
```

```python
pd_result= pd.DataFrame(fin_result)
pd_result[:55].to_csv('./conference.txt') # 2016년 12월~ 최근
```
